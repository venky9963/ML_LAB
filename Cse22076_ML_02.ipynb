```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import pandas as pd

# Reload the data
data = pd.read_excel('Lab Session Data.xlsx', sheet_name='thyroid0387_UCI')

# Convert columns to string if they contain mixed types
for col in data.select_dtypes(include=['object']).columns:
    if data[col].apply(type).nunique() > 1:
        data[col] = data[col].astype(str)

# Identify categorical columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Initialize LabelEncoder and OneHotEncoder
label_encoders = {}
one_hot_encoded_df = data.copy()

# Apply Label Encoding for ordinal variables if known
for col in categorical_cols:
    le = LabelEncoder()
    one_hot_encoded_df[col] = le.fit_transform(one_hot_encoded_df[col])
    label_encoders[col] = le

# Apply One-Hot Encoding for nominal variables
one_hot_encoded_df = pd.get_dummies(one_hot_encoded_df, columns=categorical_cols)

print(one_hot_encoded_df.head())
```

```python
#A7
#Apply Normalization Techniques
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Initialize scalers
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()

# Select numeric columns
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns

# Min-Max Normalization
data_min_max = data.copy()
data_min_max[numeric_cols] = min_max_scaler.fit_transform(data[numeric_cols])

# Standardization
data_standardized = data.copy()
data_standardized[numeric_cols] = standard_scaler.fit_transform(data[numeric_cols])

# Display results
print("Data after Min-Max Normalization:")
print(data_min_max.head())

print("Data after Standardization:")
print(data_standardized.head())
```

```python
#A8
#  Filter Binary Attributes
# Filter columns with binary values
binary_cols = data.columns[data.isin([0, 1]).all()]
binary_data = data[binary_cols]

# Display binary columns and their data
print("Binary columns:\n", binary_cols)
print(binary_data.head())
```

```python
#A8
#  Extract Binary Vectors
# Extract the first two observation vectors
vector1 = binary_data.iloc[0]
vector2 = binary_data.iloc[1]

# Display the vectors
print("First vector:\n", vector1)
print("Second vector:\n", vector2)
```

```python
#A8
#4. Compute JC and SMC
# Compute f11, f10, f01, f00
f11 = ((vector1 == 1) & (vector2 == 1)).sum()
f10 = ((vector1 == 1) & (vector2 == 0)).sum()
f01 = ((vector1 == 0) & (vector2 == 1)).sum()
f00 = ((vector1 == 0) & (vector2 == 0)).sum()

# Calculate JC and SMC
jc = f11 / (f01 + f10 + f11) if (f01 + f10 + f11) > 0 else 0
smc = (f11 + f00) / (f00 + f01 + f10 + f11) if (f00 + f01 + f10 + f11) > 0 else 0

# Display results
print(f"Jaccard Coefficient (JC): {jc}")
print(f"Simple Matching Coefficient (SMC): {smc}")
```
